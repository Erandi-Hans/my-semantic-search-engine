```python
# ==============================================================================
# 1. INSTALL REQUIRED LIBRARIES
# ==============================================================================
!pip install -q sentence-transformers pdfplumber gradio

import os
import pdfplumber
import gradio as gr
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# ==============================================================================
# 2. INITIALIZE THE SEMANTIC EMBEDDING MODEL
# ==============================================================================
print("Loading Embedding Model (all-MiniLM-L6-v2)...")
model = SentenceTransformer('all-MiniLM-L6-v2')

PDF_FOLDER = "/content/drive/MyDrive/pdfs"
all_chunks = []
corpus_embeddings = None

if not os.path.exists(PDF_FOLDER):
    os.makedirs(PDF_FOLDER)
    print(f"Created '{PDF_FOLDER}' folder. Please upload your domain sub-folders and PDFs into it.")

# ==============================================================================
# 3. PDF PROCESSING & SEMANTIC CHUNKING PIPELINE
# ==============================================================================
def process_pdfs():
    global all_chunks, corpus_embeddings
    all_chunks = []

    print("Scanning folders and processing PDFs...")

    for root, dirs, files in os.walk(PDF_FOLDER):
        for file in files:
            if file.endswith('.pdf'):
                pdf_path = os.path.join(root, file)
                domain_name = os.path.basename(root)

                try:
                    with pdfplumber.open(pdf_path) as pdf:
                        for page_num, page in enumerate(pdf.pages):
                            text = page.extract_text()
                            if text:
                                paragraphs = text.split("\n\n")
                                for para in paragraphs:
                                    if len(para.strip()) > 30:
                                        all_chunks.append({
                                            "pdf_path": pdf_path,
                                            "pdf_name": file,
                                            "domain": domain_name,
                                            "page": page_num + 1,
                                            "text": para.strip()
                                        })
                except Exception as e:
                    print(f"Error reading file {file}: {e}")

    if all_chunks:
        print(f"Extracted {len(all_chunks)} semantic text chunks.")
        corpus_texts = [chunk["text"] for chunk in all_chunks]
        print("Generating Semantic Vector Embeddings...")
        corpus_embeddings = model.encode(corpus_texts, show_progress_bar=True)
        print("Pipeline initialization complete!")
    else:
        print("System warning: No PDF files detected inside folder yet.")

process_pdfs()

# ==============================================================================
# 4. CORE SEMANTIC SEARCH LOGIC (VERTICAL FULL WIDTH LAYOUT)
# ==============================================================================
def search_and_display(query):
    if not all_chunks:
        return "<h3 style='color:red;'>Error: Please upload PDFs to the 'pdfs' folder first.</h3>", None
    if not query.strip():
        return "<h3>Please enter a search query.</h3>", None

    query_embedding = model.encode([query])
    similarities = cosine_similarity(query_embedding, corpus_embeddings)[0]

    sorted_indices = similarities.argsort()[::-1]

    # Filters out any results below 20% confidence score
    CONFIDENCE_THRESHOLD = 0.20
    valid_indices = [idx for idx in sorted_indices if similarities[idx] >= CONFIDENCE_THRESHOLD]

    if not valid_indices:
        return """
        <div style='background-color: #ffebd2; padding: 15px; border-radius: 8px; border-left: 5px solid #ff9800; font-family: sans-serif; width: 100%; box-sizing: border-box;'>
            <h3 style='margin: 0; color: #cc7a00;'>Low Confidence Match</h3>
            <p>No highly relevant semantic content was discovered for this query in the document matrix.</p>
        </div>
        """, None

    # Full Width Layout Content Box
    results_html = """
    <div style='font-family: sans-serif; width: 100%; box-sizing: border-box;'>
        <h3 style='color: #0d47a1; margin-bottom: 20px; text-align: left;'> Relevant Semantic Matches Found</h3>
    """

    for rank, idx in enumerate(valid_indices):
        score = similarities[idx]
        match = all_chunks[idx]

        badge_color = "#4caf50" if rank == 0 else "#2196f3"
        rank_label = f"Match #{rank+1}"

        # Clean vertical card structured for seamless full-width scrolling
        results_html += f"""
        <div style='background-color: #f8f9fa; padding: 20px; margin-bottom: 20px; border-radius: 8px; border-left: 5px solid {badge_color}; box-shadow: 0 2px 5px rgba(0,0,0,0.05); width: 100%; box-sizing: border-box;'>
            <table style='width: 100%; border-collapse: collapse; margin-bottom: 12px;'>
                <tr>
                    <td><span style='background-color: {badge_color}; color: white; padding: 4px 10px; border-radius: 4px; font-weight: bold; font-size: 12px;'>{rank_label}</span></td>
                    <td style='text-align: right; font-size: 13px; color: #555;'><b>Confidence Score:</b> <span style='background-color: #e0e0e0; padding: 3px 8px; border-radius: 4px; font-weight: bold;'>{score:.4f}</span></td>
                </tr>
            </table>
            <div style='font-size: 14px; color: #333; margin-bottom: 12px; line-height: 1.5;'>
                <p style='margin: 4px 0;'><b>Domain Folder:</b> <span style='color: #1565c0;'>{match['domain']}</span></p>
                <p style='margin: 4px 0;'><b>Source File:</b> {match['pdf_name']}</p>
                <p style='margin: 4px 0;'><b>Location:</b> Page {match['page']}</p>
            </div>
            <div style='background-color: #ffffff; padding: 15px; border: 1px solid #e0e0e0; border-radius: 6px; font-style: italic; color: #222222; font-size: 14px; line-height: 1.5; box-shadow: inset 0 1px 3px rgba(0,0,0,0.02);'>
                "{match['text']}"
            </div>
        </div>
        """

    results_html += "</div>"
    best_pdf_path = all_chunks[valid_indices[0]]['pdf_path']

    return results_html, best_pdf_path

# ==============================================================================
# 5. GRADIO GRAPHICAL USER INTERFACE (FULL WIDTH LAYOUT)
# ==============================================================================
custom_css = """
#main-layout-col {
    max-width: 98%;
    margin-left: auto;
    margin-right: auto;
    padding: 10px;
}
.center-text {
    text-align: center;
}
"""

with gr.Blocks(theme=gr.themes.Soft(), title="Semantic Search Engine", css=custom_css) as demo:
    with gr.Column(elem_id="main-layout-col"):
        gr.Markdown("# Cross-Domain Multi-Match Semantic Search Engine", elem_classes="center-text")
        gr.Markdown("Search your document catalog conceptually. All valid matching paragraphs are filtered and listed vertically.", elem_classes="center-text")

        # Row 1: Search Input & Button
        with gr.Row():
            query_input = gr.Textbox(
                label="Search Input Query",
                placeholder="Type conceptual ideas here...",
                scale=4
            )
            search_btn = gr.Button("Execute Semantic Search ", variant="primary", scale=1)

        # Row 2: Document Viewer (Full Width - Styled to appear above or neatly below)
        with gr.Row():
            with gr.Column(scale=12):
                gr.Markdown("### Top Match Document Viewer")
                pdf_output = gr.File(label="Suggested Source PDF File (Rank #1)")

        # Row 3: Search Results (Full Width)
        with gr.Row():
            with gr.Column(scale=12):
                meta_output = gr.HTML(label="Ranked Search Results")

        search_btn.click(
            fn=search_and_display,
            inputs=query_input,
            outputs=[meta_output, pdf_output]
        )

demo.launch(share=True, debug=True)
```
